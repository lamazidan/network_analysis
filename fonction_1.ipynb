```python
def traiter_et_afficher_auteurs_et_dimensions(df):
    # Afficher les dimensions du DataFrame original
    print(f"Dimensions originales : {df.shape}")

    # Fusionner toutes les colonnes d'auteurs en une seule Series pour le DataFrame original
    auteurs_original = pd.Series(dtype='object')
    for col in df.columns:
        if 'marcrel:aut{URIRef}' in col:
            auteurs_original = auteurs_original.append(df[col].dropna(), ignore_index=True)

    # Compter le nombre total d'auteurs uniques dans le DataFrame original
    nombre_auteurs_uniques_original = auteurs_original.nunique()
    print(f"Nombre d'auteurs uniques originaux : {nombre_auteurs_uniques_original}")

    # Identifier les auteurs les plus fréquents
    top_auteurs = auteurs_original.value_counts().nlargest(6).index

    # Filtrer les lignes correspondantes à ces auteurs dans le DataFrame
    df_top_auteurs = pd.DataFrame()
    for col in df.columns:
        if 'marcrel:aut{URIRef}' in col:
            is_in_top_auteurs = df[col].isin(top_auteurs)
            df_top_auteurs = df_top_auteurs.append(df[is_in_top_auteurs], ignore_index=True)

    # Éliminer les doublons
    df_top_auteurs = df_top_auteurs.drop_duplicates()

    # Fusionner les colonnes d'auteurs du DataFrame filtré pour compter les auteurs uniques
    auteurs_top_auteurs = pd.Series(dtype='object')
    for col in df_top_auteurs.columns:
        if 'marcrel:aut{URIRef}' in col:
            auteurs_top_auteurs = auteurs_top_auteurs.append(df_top_auteurs[col].dropna(), ignore_index=True)
    nombre_auteurs_uniques_top = auteurs_top_auteurs.nunique()

    print(f"Dimensions après réduction : {df_top_auteurs.shape}")
    print(f"Nombre d'auteurs uniques dans le DataFrame réduit : {nombre_auteurs_uniques_top}")

    return df_top_auteurs
```

```python
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9]
dfs_reduits = []  # Liste pour stocker les DataFrame réduits

for i, df in enumerate(dfs, start=1):
    print(f"\n--- Traitement de DataFrame {i} ---")
    df_reduit = traiter_et_afficher_auteurs_et_dimensions(df)
    dfs_reduits.append(df_reduit)
```
